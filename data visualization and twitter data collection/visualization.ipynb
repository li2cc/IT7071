There are programming packages and tools developed for the very purpose. 
[Matplotlib](http://matplotlib.org) provides a comprehensive selection of chart types for data science. [Seaborn](http://seaborn.pydata.org/) and [ggplot](http://ggplot.yhathq.com/), both built on top of Matplotlib, provides enhanced presentation of diagrams, purticularly embedded statistical analysis. We will learn some basic usages of both Matplotlib and seaborn. 


```python
#import pyplot subpackage from matplotlib as plt
from matplotlib import pyplot as plt
#make diagram display within notebook
%matplotlib inline
#we also need numpy to convert data in array foramt 
import numpy as np
```

```python
#use random to generate the dataset (x, y)
import random
#set a seed, so we generate the same random dataset
random.seed(0)
#x follows normal distributuion with a mean of 2000, std 100, and 200 data points
x = np.random.normal(2000, 100, 200)
#y is the sum of x and noise 
y = (x + np.random.normal(0,400,200))
```

```python
#let's try a basic plot (line chart)
plt.plot(x,y)
```

```python
#scatter plot may make more sense of the data than line chart
plt.scatter(x,y)
```

```python
#seaborn has an enhanced scatter chart based on matplotlib
#seaborn jointplot
import seaborn as sns
#there are options to see data distribution density 
#let's try "hex". Use Shift+Tab to see other options
#online documentations and examples from user community are the best references form
#learning different drawing options.
sns.jointplot(x,y, kind='hex' )
```

```python
#We know that x and y has a linear relationship 
#because that is how we created y :-) 
#let's sort x and y in ascending order 
x.sort()
y.sort()
x
```

```python
#let's plot the line char again
#now, the linear relationship between x and y is seen
plt.plot(x,y)
```

```python
import seaborn as sns
```

```python
#we use joinplot again
#explain the relationship here
sns.jointplot(x,y,kind='hex')
```

```python
#histogram is the best choice for seeing data distribution 
#the shape of the histogram indicates data distribution type
plt.hist(x)
```

```python
#we can draw two charts on one diagram
plt.hist(x)
plt.hist(y)
```

```python
#let's play with some parameters to make the diagram look better
plt.hist(x, normed=True, color='r', bins=10)
plt.hist(y, normed=True, alpha=0.6, bins=20)
plt.show()
```

```python
#seaborn distplot is an enhanced histogram 
sns.distplot(x, norm_hist=True, kde=False)
sns.distplot(y, norm_hist=True)
```

```python
#boxplot is good for seeing standard deviation and outliers
sns.boxplot(data=[x,y], orient='horizontal')
```

```python
#vilinplot further sees the distribution of data within the "box"
sns.violinplot(data=[x,y], )
```

```python
#let's use pandas to create a data frame
#we will learn pandas in the future. At this point of time,
#you only need to that our dataset has two columns x, y
import pandas as pd
df = pd.DataFrame()
df['x'] =x
df['y'] =y

#Let's use the seaborn statistical analysis capacity to find 
#the linear relationship of x and y
#lmplot find the best fitting line
sns.lmplot(data = df, x='x', y='y', order=4, palette="Set1" )
```
